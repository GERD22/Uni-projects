```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
df = pd.read_csv("synthetic.csv").drop(['Unnamed: 0'], axis=1)

df.shape
print(df)
# Do not forget to use the scaling
X_std = StandardScaler().fit_transform(df.values)

# Determine the number of main components.
nr_components=3
print(X_std[:3,:])
```

```python
%%time
class PCS:
    
    
    def svd(self, x):
        return np.linalg.eig(np.matmul(x, x))
        
    def __init__(self,n_components=None,method="svd"):
        self.n_components=n_components
        self.method=method
    def explained_ratio(self):
        
        return np.sum(self.v[:self.n_components]) / np.sum(self.v)
        

    def fit_transform(self, df):
        df = df - np.mean(df, axis=0)
        cov = np.cov(np.transpose(df))
        v, U = self.svd(cov)
        self.v = v
        U_reduced = U[:, :self.n_components]
        X_reduced = np.matmul(df, U_reduced)
        return X_reduced

model = PCS(n_components=2)
print(model.fit_transform(np.array(df)))

                
    
        
```
